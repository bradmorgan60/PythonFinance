Predict Tesla's Gross Profit

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 29590 # expected revenue 2020 (in millions)
rev_std = 30 # standard deviation of revenue
iterations = 1000 # number of simulations of the company's expected revenues
```

```python
rev = np.random.normal(rev_m, rev_std, iterations)
rev
# produces simulation of future revenues
# there will be 1000 values
```

```python
plt.figure(figsize=(15,5))
plt.plot(rev)
plt.show()

# output centers around the mean of 170
# since the standard deviation is 20, the chart fluctuates on average between 150 and 190
# 170 + 20 and 170 - 20
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))
# COGS is negative becasue it is money spent
# In this simulation, I am assuming COGS to be about 60% of revenues --> 0.6
# The distribution deviates 10% regarding the mean of COGS --> 0.1

# revenue value we obtained above must be extracted from a random normal distribution


plt.figure(figsize=(15,5))
plt.plot(COGS)
plt.show()
```

```python
Gross_Profit = rev + COGS
Gross_Profit
```
