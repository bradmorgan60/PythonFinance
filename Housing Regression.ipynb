```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("C:\PythonFF\Housing.xlsx")
```

```python
data
```

Regression

Step 1: Locate indepedent variables and dependent variables

```python
X = data[['House Size (sq.ft.)','Number of Rooms','Year of Construction']]
# X values == independent variables
Y = data['House Price']
# House price is the dependent variable
```

```python
X1 = sm.add_constant(X)
reg = sm.OLS(Y, X1).fit()

reg.summary()

# R-squared showcases the model's explanatory power aka if the model is any good or not
# in this case, the model shows that there is a correlation between the house attributes and house price
```
