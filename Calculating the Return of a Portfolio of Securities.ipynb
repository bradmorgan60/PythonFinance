```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['HD', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo', start='2005-01-01')['Adj Close']
```

```python
returns = (mydata / mydata.shift(1)) - 1
returns.head()
```

```python
weights = np.array(['.1246','.1770','.2742','.1745','.2497'])
```

```python
annual_returns = returns.mean() * 250
print(annual_returns)
```

```python
print(np.dot(annual_returns, weights))
```

```python
portfolio_1 = str(round(np.dot(annual_returns, weights),5)*100) +'%'
print(portfolio_1)
```
